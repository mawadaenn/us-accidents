```python
import pandas as pd
```

```python
file = 'US_Accidents.csv'
df = pd.read_csv(file)
```

```python
df['Temperature(F)'].isna().sum()
```

```python
#transforme Weather_Timestamp (qui est un texte / object), en vraie date-heure exploitable
df['Weather_Timestamp']= pd.to_datetime(df['Weather_Timestamp'])
#supprime l’heure,garde uniquement la date (YYYY-MM-DD)
df['Weather_Date']=df['Weather_Timestamp'].dt.date
```

```python
# 1. Date + City + State + Street
df['Temperature(F)'] = df['Temperature(F)'].fillna(
    df.groupby(['Weather_Date','State','City','Street'])['Temperature(F)'].transform('median')
)
df['Temperature(F)'].isna().sum()
```

```python
# 2. Date + State + City
df['Temperature(F)'] = df['Temperature(F)'].fillna(
    df.groupby(['Weather_Date','State','City'])['Temperature(F)'].transform('median')
)
df['Temperature(F)'].isna().sum()
```

```python
# 3. Date + State
df['Temperature(F)'] = df['Temperature(F)'].fillna(
    df.groupby(['Weather_Date', 'State'])['Temperature(F)'].transform('median')
)
df['Temperature(F)'].isna().sum()
```

```python
# 4. Médiane globale
df['Temperature(F)'] = df['Temperature(F)'].fillna(df['Temperature(F)'].median())
df['Temperature(F)'].isna().sum()
```

```python
#Date + State
df['Humidity(%)'] = df['Humidity(%)'].fillna(
    df.groupby(['Weather_Date', 'State'])['Humidity(%)']
      .transform('median')
)

df['Humidity(%)'].isna().sum()
```

```python
#Médiane globale (fallback final)
df['Humidity(%)'] = df['Humidity(%)'].fillna(df['Humidity(%)'].median())

df['Humidity(%)'].isna().sum()
```

```python
df['Weather_Condition'].isna().sum()
```

```python
# Colonnes pour le groupby
cols_group = ['Weather_Date', 'State', 'City', 'Street']
```

```python
#Calcul du mode de Weather_Condition par groupe
mode_per_group = (
    df
    .dropna(subset=['Weather_Condition'])
    .groupby(cols_group)['Weather_Condition']
    .agg(lambda x: x.mode().iloc[0])
    .reset_index()
    .rename(columns={'Weather_Condition': 'Weather_Condition_mode'})
)
```

```python
#Jointure pour récupérer le mode
df = df.merge(
    mode_per_group,
    on=cols_group,
    how='left'
)

# Remplissage des NaN par le mode du groupe
df['Weather_Condition'] = df['Weather_Condition'].fillna(df['Weather_Condition_mode'])

#Suppression de la colonne temporaire
df = df.drop(columns=['Weather_Condition_mode'])

#Remplissage de sécurité par le mode global
global_mode = df['Weather_Condition'].mode()[0]
df['Weather_Condition'] = df['Weather_Condition'].fillna(global_mode)

#Vérification finale
print("Nombre de NaN restants :", df['Weather_Condition'].isna().sum())
```
