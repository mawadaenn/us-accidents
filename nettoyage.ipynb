```python
#pour remplis les valeur manquants dans la collonne wind_chill et donnees la justification
import pandas as pd
file = 'US_Accidents.csv'
df = pd.read_csv(file)
```

```python
# Remplir les valeurs manquantes dans la colonne 'Sunrise_Sunset' en fonction de temps si >= 00:00 AM alors 'Day' >= 06:00 PM alors 'Night'
for index, row in df.iterrows():
    if pd.isna(row['Sunrise_Sunset']):
        time_str = row['Start_Time'].split(' ')[1]  # Extraire l'heure de la cha√Æne Start_Time
        hour = int(time_str.split(':')[0])
        if 0 <= hour < 6 or hour >= 18:
            df.at[index, 'Sunrise_Sunset'] = 'Night'
        else:
            df.at[index, 'Sunrise_Sunset'] = 'Day'  
        df.at[index, 'sunrise_sunset_justification'] = f"Filled based on time {time_str}"
```

```python
# pour la supprission des colonnes wind_chill, astromical_twilight, nautical_twilight, civil_twilight
df = df.drop(columns=['Wind_Chill(F)', 'Astronomical_Twilight', 'Nautical_Twilight', 'Civil_Twilight']) 
```
