```python
#pour remplis les valeur manquants dans la collonne wind_chill et donnees la justification
import pandas as pd
import numpy as np
import math
# Charger les données depuis le fichier CSV
df = pd.read_csv('Us_Accidents.csv')
# Fonction pour calculer le wind chill
def calculate_wind_chill(temp_f, wind_mph):
    if temp_f is None or wind_mph is None or math.isnan(temp_f) or math.isnan(wind_mph):
        return None
    if temp_f > 50 or wind_mph < 3:
        return temp_f
    wind_chill = 35.74 + 0.6215 * temp_f - 35.75 * (wind_mph ** 0.16) + 0.4275 * temp_f * (wind_mph ** 0.16)
    return round(wind_chill, 2)
# Remplir les valeurs manquantes dans la colonne 'wind_chill'
for index, row in df.iterrows():
    if pd.isna(row['wind_chill']):
        temp_f = row['temperature(F)']
        wind_mph = row['wind_speed(mph)']
        wind_chill_value = calculate_wind_chill(temp_f, wind_mph)
        df.at[index, 'wind_chill'] = wind_chill_value
        df.at[index, 'wind_chill_justification'] = f"Calculated using temperature {temp_f}F and wind speed {wind_mph}mph"
# Enregistrer les données mises à jour dans un nouveau fichier CSV
df.to_csv('Us_Accidents_Updated.csv', index=False)
# justification: Les valeurs manquantes dans la colonne 'wind_chill' ont été remplies en utilisant la formule standard du wind chill basée sur la température et la vitesse du vent. Si les conditions pour le calcul du wind chill ne sont pas remplies (température > 50°F ou vitesse du vent < 3 mph), la température elle-même est utilisée comme valeur de wind chill.
```
