```python
import pandas as pd

# Conversion des colonnes en datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'], errors='coerce')

# Vérification du type
df[['Start_Time', 'End_Time']].dtypes
```

# Qualité des valeurs manquantes 

```python
missing_count = df.isna().sum()
missing_percent = (df.isna().mean() * 100).round(2)

missing_df = pd.DataFrame({
    'Missing_Count': missing_count,
    'Missing_%': missing_percent
}).sort_values('Missing_%', ascending=False)

missing_df
```

# Analyse spécial (ville)

```python
top_10_cities = (
    df['City']
    .value_counts()
    .head(10)
    .reset_index()
)

top_10_cities.columns = ['City', 'Nombre_Accidents']
top_10_cities
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=top_10_cities,
    x='Nombre_Accidents',
    y='City'
)
plt.title("Top 10 des villes avec le plus d'accidents")
plt.xlabel("Nombre d'accidents")
plt.ylabel("Ville")
plt.show()
```

# Analyse temporelle

```python
df['Month'] = df['Start_Time'].dt.month
df['Month_Name'] = df['Start_Time'].dt.month_name()

df['DayOfWeek'] = df['Start_Time'].dt.day_name()

df['Hour'] = df['Start_Time'].dt.hour
df
```

```python
accidents_by_month = df['Month_Name'].value_counts()

mois_max = accidents_by_month.idxmax()
nb_mois_max = accidents_by_month.max()

mois_max, nb_mois_max
```

```python
accidents_by_day = df['DayOfWeek'].value_counts()

jour_max = accidents_by_day.idxmax()
nb_jour_max = accidents_by_day.max()

jour_max, nb_jour_max
```

```python
df['Year'] = df['Start_Time'].dt.year
```

```python
accidents_by_hour = df['Hour'].value_counts().sort_index()

heure_max = accidents_by_hour.idxmax()
nb_heure_max = accidents_by_hour.max()

heure_max, nb_heure_max
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    x=accidents_by_month.index,
    y=accidents_by_month.values
)
plt.xticks(rotation=45)
plt.title("Accidents par mois")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    x=accidents_by_day.index,
    y=accidents_by_day.values
)
plt.xticks(rotation=45)
plt.title("Accidents par jour")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    x=accidents_by_hour.index,
    y=accidents_by_hour.values
)
plt.xticks(rotation=45)
plt.title("Accidents par heure")
plt.show()
```

# Serivité des accidents 

```python
severity_counts = df['Severity'].value_counts().sort_index()
severity_counts
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=severity_counts.index, y=severity_counts.values)
plt.title("Distribution de la gravité des accidents")
plt.xlabel("Niveau de sévérité")
plt.ylabel("Nombre d'accidents")
plt.show()
```

```python
severity_weather = (
    df.groupby('Weather_Condition')['Severity']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    
)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=severity_weather.values, y=severity_weather.index)
plt.title("Sévérité moyenne par condition météo")
plt.show()
```
