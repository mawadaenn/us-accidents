# Nettoyage des donneés 

### remplire la colonne city 

```python
import pandas as pd 
```

```python
data=pd.read_csv("US_Accidents.csv")
```

```python


# Exemple : df est ton DataFrame

# 1. Créer un mapping Zipcode → City (la plus fréquente)
zip_city_map = (
    data.dropna(subset=["City"])
      .groupby("Zipcode")["City"]
      .agg(lambda x: x.mode().iloc[0])
)

# 2. Remplir les City manquantes
data["City"] = data["City"].fillna(data["Zipcode"].map(zip_city_map))
```

```python
data["City"].isna().sum()
```

### remplire la colonne Zipcode 

```python
data["Zipcode"].isna().sum()
```

```python
# 1. Mapping City → Zipcode (le plus fréquent)
city_zip_map = (
    data.dropna(subset=["Zipcode"])
      .groupby("City")["Zipcode"]
      .agg(lambda x: x.mode().iloc[0])
)

# 2. Remplir les Zipcode manquants
data["Zipcode"] = data["Zipcode"].fillna(data["City"].map(city_zip_map))
```

### remplire la colonne Timezone

```python
data["Timezone"].isna().sum()
```

```python
# 1. Mapping State → TimeZone (la plus fréquente)
state_tz_map = (
    data.dropna(subset=["Timezone"])
      .groupby("State")["Timezone"]
      .agg(lambda x: x.mode().iloc[0])
)

# 2. Remplir les Timezone manquantes
data["Timezone"] = data["Timezone"].fillna(data["State"].map(state_tz_map))
```

```python
data["Timezone"].isna().sum()
```
